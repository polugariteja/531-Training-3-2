**Softmax vs Sigmoid (Comparison)**

```python
import numpy as np

logits = np.array([2.0, 1.0, 0.1])

sigmoid = 1 / (1 + np.exp(-logits))

exp_logits = np.exp(logits)
softmax = exp_logits / np.sum(exp_logits)

print("Logits:", logits)
print("Sigmoid Output:", sigmoid)
print("Sum of Sigmoid outputs:", np.sum(sigmoid))
print()
print("Softmax Output:", softmax)
print("Sum of Softmax outputs:", np.sum(softmax))
```

**Softmax + Cross Entropy Loss**

```python
import numpy as np

logits = np.array([2.0, 1.0, 0.1])
true_class = 0

exp_logits = np.exp(logits)
softmax = exp_logits / np.sum(exp_logits)

loss_manual = -np.log(softmax[true_class])

loss_numpy = -np.sum(np.eye(len(logits))[true_class] * np.log(softmax))

print("Softmax:", softmax)
print("Manual Loss:", loss_manual)
print("NumPy Loss:", loss_numpy)
```

**Visualizing Softmax Output**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
logits = np.random.randn(5)

exp_logits = np.exp(logits)
softmax = exp_logits / np.sum(exp_logits)

plt.figure()
plt.bar(range(len(softmax)), softmax)
plt.xlabel("Class Index")
plt.ylabel("Probability")
plt.title("Softmax Probability Distribution")
plt.show()

```

**Mini-Project â€“ Softmax in Language Models**

```python
import numpy as np

vocab = ["I", "like", "machine", "translation"]

def softmax(logits):
    exp_logits = np.exp(logits)
    return exp_logits / np.sum(exp_logits)

def greedy_decode(probs):
    return np.argmax(probs)

def random_sample(probs):
    return np.random.choice(len(probs), p=probs)

np.random.seed(0)

sentence_greedy = []
sentence_random = []

for step in range(3):
    logits = np.random.randn(len(vocab))
    probs = softmax(logits)

    g_idx = greedy_decode(probs)
    r_idx = random_sample(probs)

    sentence_greedy.append(vocab[g_idx])
    sentence_random.append(vocab[r_idx])

    print(f"Step {step+1}")
    print("Logits:", logits)
    print("Probabilities:", probs)
    print("Greedy word:", vocab[g_idx])
    print("Random word:", vocab[r_idx])
    print()

print("Greedy Sentence:", " ".join(sentence_greedy))
print("Random Sentence:", " ".join(sentence_random))
```
